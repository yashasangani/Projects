### THEREFORE HERE OUR TARGET IS TO FIND M AND C USING GRADIENT DESCENT APPROACH AND THEN USING THAT M AND C WE WILL PREDICT ON TEST DATASET.

```python
x_train = df.drop(' Y',axis=1)
y_train = df[" Y"]
x_train.shape
```

```python
dtest = pd.read_csv("https://raw.githubusercontent.com/yashasangani/Datasets/main/0000000000002417_test_boston_x_test.csv")
dtest.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test= scaler.transform(dtest)
```

```python
def gradient_descent(x,y,iterations,learning_rate):
  m = np.zeros(x.shape[1])
  c =0
  n = len(x)
  for i in range(iterations):
    md = (-2/n)*(((y - (x*m).sum(axis=1) - c).values.reshape(-1, 1))*x).sum(axis=0)
    cd = (-2/n)*(((y - (x*m).sum(axis=1)-c).values.reshape(-1, 1))).sum(axis=0)
    m = m - learning_rate * md
    c = c - learning_rate * cd
    print("iterations =", i ,"cost = ",cost_function(x,y,iterations,m,c))
  return m,c
```

```python
def cost_function(x,y,iterations,m,c):
  for i in range(iterations):
    cost = ((y - (x*m).sum(axis=1)-c) ** 2).mean()
    return cost
```

```python
def run(x,y):
  m = np.zeros(x.shape[1])
  c =0
  n = len(x)
  learning_rate = 0.009
  iterations = 1000

  m,c = gradient_descent(x,y,iterations,learning_rate)
  return m[0:13],c



```

```python
m,c = run(x_train,y_train)
print(f"m = {m[0:13]} \nc = {c}")

```

```python
def predict(x_train,m,c):
    n=len(x_train)
    y_pred=np.zeros(n)
    for i in range(n):
        X=x_train[i,:]
        y_pred[i]=(m*X).sum()+c

    return y_pred
```

```python
y_pred = predict(x_train,m,c)
```
